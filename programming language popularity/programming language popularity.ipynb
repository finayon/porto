```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"QueryResults.csv", names = ['DATE', 'TAG', 'POSTS'], header = 0)
df
```

```python
#adjust the date
pd.to_datetime(df['DATE'])
```

```python
#Which programming language had the most number of posts since the creation of Stack Overflow?
df.groupby('TAG').sum()
```

```python
#how many months of entries exist per programming language.
df.groupby('TAG').count()
```

```python
reshaped_df = df.pivot(index='DATE', columns='TAG', values='POSTS')
reshaped_df
```

```python
reshaped_df.fillna(0, inplace = True)
reshaped_df.isna().values.any()
```

```python
plt.figure(figsize=(16, 10))
plt.xticks(range(0, len(reshaped_df.index) + 1, 25))
plt.plot(reshaped_df.index, reshaped_df.java)
```
